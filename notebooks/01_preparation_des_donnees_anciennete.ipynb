```python
import pandas as pd
import numpy as np
import os
from pprint import pprint
```

```python
path_to_raw_data = os.path.join('..', 'raw_data')
print(path_to_raw_data)
```

```python
list_of_files = sorted(os.listdir(path_to_raw_data))
pprint(list_of_files)
```

```python
df = pd.read_excel(
    os.path.join(path_to_raw_data, list_of_files[0]),
    sheet_name='par ancienneté',
    skiprows=3,
    skipfooter=2,
    engine='openpyxl'
)

df = df.iloc[:, :8]

df.columns = [
    'lib_epci',
    'anc_moins_3_mois',
    'anc_3_a_5_mois',
    'anc_6_a_11_mois',
    'anc_12_a_23_mois',
    'anc_24_a_35_mois',
    'anc_plus_36_mois',
    'total'
]
```

```python
df.head()
```

```python
df.tail()
```

```python
df.to_csv(
    os.path.join(
        '..',
        'clean_data',
        'anciennete',
        '2017.csv'
    ),
    index=False
)
```
