```python
import pandas as pd
import numpy as np
import os
from pprint import pprint
```

```python
path_to_raw_data = os.path.join('..', 'raw_data')
print(path_to_raw_data)
```

```python
list_of_files = sorted(os.listdir(path_to_raw_data))
pprint(list_of_files)
```

```python
df = pd.read_excel(
    os.path.join(path_to_raw_data, list_of_files[0]),
    sheet_name='ABCpar_age_sexe_diplome',
    skiprows=3,
    skipfooter=2,
    engine='openpyxl'
)

df = df.dropna(axis=1, how='all')

df.columns = [
    'lib_epci',
    'hommes_moins_de_25_ans',
    'hommes_25_a_49_ans',
    'hommes_plus_de_50_ans',
    'total_hommes',
    'femmes_moins_de_25_ans',
    'femmes_25_a_49_ans',
    'femmes_plus_de_50_ans',
    'total_femmes',
    'diplome_niv_1_2',
    'diplome_niv_3',
    'diplome_niv_4',
    'diplome_niv_5',
    'diplome_niv_5_bis',
    'diplome_niv_6',
    'diplome_indetermine',
    'total_categorie_abc'
]

df.head()
```

```python
df.tail()
```

```python
df.to_csv(
    os.path.join(
        '..',
        'clean_data',
        'categorie_abc',
        '2017.csv'
    ),
    index=False
)
```
