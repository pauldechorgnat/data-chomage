```python
import pandas as pd
import numpy as np
import os
from pprint import pprint
```

```python
path_to_raw_data = os.path.join('..', 'raw_data', 'epci')
print(path_to_raw_data)
```

```python
list_of_files = sorted(os.listdir(path_to_raw_data))
pprint(list_of_files)
```

## Traitement des données de 2016

> Les communautés urbaines (CU), communautés d'agglomération (CA), communautés de communes (CC), syndicats d'agglomération nouvelle (SA), et  métropoles (ME) sont les EPCI à fiscalité propre retenus ici.

```python
liste_epci = pd.read_excel(
    os.path.join(path_to_raw_data, list_of_files[1]),
    sheet_name='Liste_des_EPCI',
    skiprows=5,
#     engine='openpyxl',
)

liste_epci.head()
```

```python

liste_epci.tail()
```

```python
composition_communale = pd.read_excel(
    os.path.join(path_to_raw_data, list_of_files[1]),
    sheet_name='Composition_communale',
    skiprows=5,
#     engine='openpyxl',
)
```

```python
composition_communale.head()
```

```python
liste_epci.tail()
```

```python
liste_epci.columns = [
    'code_epci', 
    'libelle_epci',
    'nature_epci',
    'nb_communes'
]
```

```python
len(composition_communale)
```

```python
composition_communale.head()
```

```python
composition_communale.info()
```

```python
composition_communale.columns = [
    'codgeo',
    'libgeo',
    'code_epci',
    'libelle_epci',
    'code_departement',
    'code_region'
]
```

```python
df = pd.merge(
    left=composition_communale,
    right=liste_epci,
    how='inner',
    on=['code_epci', 'libelle_epci']
)
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
verif1 = df.groupby(['code_epci']).count()['nb_communes'].reset_index()
verif1.head()
```

```python
verif2 = pd.merge(left=verif1, right=liste_epci, how='outer')

verif2
```

```python
verif1.shape
```

```python
liste_epci.shape
```

```python
liste_epci
```

```python
df.head()
```

```python
df.to_csv(
    os.path.join(
        '..',
        'clean_data',
        'epci',
        '2016.csv'
    ),
    index=False
)
```
